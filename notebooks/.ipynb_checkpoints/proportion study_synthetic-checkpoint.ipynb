```python

def plot_(x,y,ylabel,xlabel,title):
    f = plt.figure(figsize=(4,4))
    ax = f.add_subplot(111)
    ax.scatter(x, y, color = 'black',s=3)
    
    #ax.plot(x,y)
    ax.grid(False)
    
    # Only show ticks on the left and bottom spines
    ax.yaxis.set_ticks_position('left')
    ax.xaxis.set_ticks_position('bottom')
    plt.xlabel(xlabel, fontsize=13)
    plt.ylabel(ylabel, fontsize=13)
    ax.tick_params(labelsize=12)

    #ax.set_aspect('equal', 'box')
    #ax.xaxis.set_ticklabels([])
    #ax.yaxis.set_ticklabels([])
    #plt.title(title)
    #ax.text(0.8, 0.1,'p='+str(p), ha='center', va='center', transform=ax.transAxes)
    plt.savefig(title+'.png')
    return
```

```python
def plot_r_p(x,r,p,xlabel,title,label):
    f = plt.figure(figsize=(5,5))
    ax1 = f.add_subplot(111)
    ax1.grid(False)

    color = 'green'
    ax1.set_xlabel(xlabel, fontsize=15)
    if label:
        ax1.set_ylabel('r_value', color=color, fontsize=15)
    
    ax1.plot(x, r, color=color,marker = 'o', label="r_value")
    ax1.tick_params(axis='y', labelcolor=color)
    if label:
        plt.legend(loc="upper left", fontsize=15)

    ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
    ax2.grid(False)
    color = 'b'
    if label:
        ax2.set_ylabel('p_value', color=color, fontsize=15)  # we already handled the x-label with ax1
    ax2.plot(x, p, color=color,marker = 'x', label="p_value")
    ax2.tick_params(axis='y', labelcolor=color)
    
    ax1.tick_params(labelsize=12)
    ax2.tick_params(labelsize=12)
    #fig.tight_layout()  # otherwise the right y-label is slightly clipped
    
    if label:
        plt.legend(loc="upper right", fontsize=15)
    #plt.show()
    plt.savefig(title+'.pdf')
    return
```

```python
import matplotlib.pyplot as plt
sc.set_figure_params(dpi=100, dpi_save=100, vector_friendly=True)
params = {'legend.fontsize': 8,
          'figure.figsize': (3, 3),
         'axes.labelsize':8,
         'axes.titlesize':8,
         'xtick.labelsize':8,
         'ytick.labelsize':8}
plt.rcParams.update(params)
```
